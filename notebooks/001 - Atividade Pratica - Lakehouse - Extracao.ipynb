```python
import pandas as pd
from pymongo import MongoClient

dbutils.widgets.text("mongo_uri", "")

uri = dbutils.widgets.get("mongo_uri")

db_name = "seguradora"
caminho_landing = "/Volumes/workspace/landing/dados"

tabelas = ["apolice", "carro", "cliente", "endereco", "estado", "marca", "modelo", "municipio", "regiao", "sinistro", "telefone"]
```

```python
try:
    client = MongoClient(uri)
    db = client[db_name]
    print("🚀 Iniciando extração\n")

    for col_name in tabelas:
        try:
            cursor = db[col_name].find()
            df_pandas = pd.DataFrame(list(cursor))

            if not df_pandas.empty:
                if '_id' in df_pandas.columns:
                    df_pandas = df_pandas.drop(columns=['_id'])
                
                df_pandas.to_csv(f"{caminho_landing}/{col_name}.csv", index=False, sep=',', encoding='utf-8')
                
                print(f"✅ {col_name}.csv gerado com sucesso!")
            else:
                print(f"⚠️ {col_name} estava vazia.")

        except Exception as e:
            print(f"❌ Erro em {col_name}: {e}")

    print("\n🏁 Agora os arquivos estão no formato que o script do professor pede!")

except Exception as e:
    print(f"💥 Erro de conexão: {e}")
```
